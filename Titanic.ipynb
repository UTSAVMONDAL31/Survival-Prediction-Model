```python
train_1 = train.dropna()
train_1
```

```python
test['Age'].isnull()
```

```python
sum(test['Age'].isnull())
```

```python
test_1 = test.dropna()
test_1
```

```python
x_train = train_1[['Age']]
y_train = train_1[['Survived']]
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
x_test = test_1[['Age']]
```

```python
prdt = model.predict(x_test)
prdt
```

```python
table = test_1.copy()
table['Prediction'] = prdt
```

```python
table
```

```python
table.to_csv("Prediction.csv")
```
